```python
#1. Plot the PMF and CDF of Binomial distribution(using the expression) with different 
#values of "p" for a particular value of "n" and repeat for different values of "n". 
#Verify your results using available functions in numpy, scipy and sklearn. 
```

```python
#code by Advika L
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Define parameters
n = 10
p_values = [0.1, 0.3, 0.5, 0.7, 0.9]

# Plot PMF and CDF for each value of p
fig, axs = plt.subplots(len(p_values), 2, figsize=(10, 20))
for i, p in enumerate(p_values):
    # Calculate PMF and CDF using the expressions
    x = np.arange(0, n+1)
    pmf = stats.binom.pmf(x, n, p)
    cdf = stats.binom.cdf(x, n, p)

    # Plot PMF
    axs[i, 0].bar(x, pmf)
    axs[i, 0].set_title(f"p={p}")
    axs[i, 0].set_xlabel("x")
    axs[i, 0].set_ylabel("P(X=x)")

    # Plot CDF
    axs[i, 1].step(x, cdf)
    axs[i, 1].set_title(f"p={p}")
    axs[i, 1].set_xlabel("x")
    axs[i, 1].set_ylabel("P(Xâ‰¤x)")

plt.tight_layout()
plt.show()
```

```python
#code by Advika L
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# define the parameters
n = 10
p_values = [0.2, 0.5, 0.8]

# calculate the PMF and CDF for each value of p
for p in p_values:
    # calculate the PMF using the binomial distribution formula
    pmf = [stats.binom.pmf(k, n, p) for k in range(n+1)]
    
    # calculate the CDF using the cumulative distribution function (CDF)
    cdf = [stats.binom.cdf(k, n, p) for k in range(n+1)]
    
    # plot the PMF and CDF
    fig, ax = plt.subplots(1, 2, figsize=(10, 4))
    ax[0].bar(range(n+1), pmf)
    ax[0].set_xlabel('k')
    ax[0].set_ylabel('P(X=k)')
    ax[0].set_title('PMF for p = {}'.format(p))
    
    ax[1].step(range(n+1), cdf)
    ax[1].set_xlabel('k')
    ax[1].set_ylabel('P(X<=k)')
    ax[1].set_title('CDF for p = {}'.format(p))
    
    plt.tight_layout()
    plt.show()
```

```python
#3. Plot the pdf and cdf for a Normal distribution with a particular mean and different standard deviations on the same graph.
```

```python
#code by Advika L
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Define parameters
mu = 0
sigma_values = [0.5, 1, 1.5, 2, 2.5]

# Plot PDF and CDF for each value of sigma
x = np.linspace(-10, 10, 1000)
fig, ax = plt.subplots(figsize=(10, 6))
for sigma in sigma_values:
    # Calculate PDF and CDF using the normal distribution
    pdf = stats.norm.pdf(x, mu, sigma)
    cdf = stats.norm.cdf(x, mu, sigma)

    # Plot PDF and CDF
    ax.plot(x, pdf, label=f"sigma={sigma}")
    ax.plot(x, cdf, label=f"sigma={sigma}", linestyle="--")

# Set plot labels and legend
ax.set_xlabel("x")
ax.set_ylabel("f(x) / F(x)")
ax.set_title("PDF and CDF for Normal Distribution")
ax.legend()

plt.show()
```

```python
#4. Compute the following probabilities for a RV 'X' having Normal distribution

#(i) Prob(m-sigma<X<m+sigma), Prob(m-2sigma<X<m+2sigma), Prob(m-3sigma<X<m+3sigma), where m is the mean and sigma is the SD of the distribution.

#(ii) Prob(X>N/2), where 'X' is a RV having Normal distribution with mean 'N' and any reasonable SD where 'N' is the last two digits of your Roll Number.
```

```python
#code by Advika L
import scipy.stats as stats

# Define mean and standard deviation
mu = 8
sigma = 2

# (i) Prob(m-sigma<X<m+sigma), Prob(m-2sigma<X<m+2sigma), Prob(m-3sigma<X<m+3sigma)
for k in range(1, 4):
    prob = stats.norm.cdf(mu + k*sigma, loc=mu, scale=sigma) - stats.norm.cdf(mu - k*sigma, loc=mu, scale=sigma)
    print(f"Prob(m-{k}sigma < X < m+{k}sigma): {prob:.4f}")

# (ii) Prob(X > N/2), where N=8
N = 8
prob = 1 - stats.norm.cdf(N/2, loc=mu, scale=sigma)
print(f"Prob(X > {N}/2): {prob:.4f}")
```

```python
#5. Generate a Normal random vector having 10000 values  using available functions in numpy, scipy and sklearn and plot its histogram.  
#Also convert the vector into a standard Normal distribution.
```

```python
#code by Advika L
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
from sklearn.preprocessing import StandardScaler

# Define parameters
mu, sigma = 0, 1
n = 10000

# Generate Normal random vector
x = np.random.normal(mu, sigma, n)

# Plot histogram of x
plt.hist(x, bins=50, density=True)
plt.title("Histogram of Normal Random Vector")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()

# Convert x to a standard Normal distribution
scaler = StandardScaler()
x_standard = scaler.fit_transform(x.reshape(-1, 1)).flatten()

# Plot histogram of x_standard
plt.hist(x_standard, bins=50, density=True)
plt.title("Histogram of Standard Normal Random Vector")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

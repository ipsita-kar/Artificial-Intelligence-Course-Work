```python
from sklearn.datasets import load_digits #loading the toy dataset of digits
from sklearn.feature_selection import SelectKBest, chi2
X, y = load_digits(return_X_y=True) 
X.shape #calculates shape of matrix
```

```python

X_new = SelectKBest(chi2, k=20).fit_transform(X, y)
X_new.shape
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import max_error
y_tru=[59054.92,59116.91,59093.36,59127.50]
y_pred=[59054.00,59061.23,59082.43,59089.89]
print("MSE is ",mean_squared_error(y_tru,y_pred))
print("MAE is ",mean_absolute_error(y_tru,y_pred))
print("Maximum error is ",max_error(y_tru,y_pred))
```

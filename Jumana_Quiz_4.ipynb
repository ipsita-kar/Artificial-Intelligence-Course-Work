2. If the distributions are skewed convert them to approximately normal using box cos transformation.

```python
#This code is done by Jumana
def boxcox(x,lmbda):
  if lmbda==0:
    return np.log(x)
  else:
    return (np.power(x,lmbda)-1)/lmbda
x=data
lmbda=316411  #taking lambda value from the PF column 
normal=boxcox(x,lmbda)
print(normal)
```

3.  Find all the descriptive parameters(mean, median, mode, variance, sd, IQR, skew and kurtosis) discussed in the class for all the 6 variables.  

**MEAN:**

```python
#This code is done by Jumana
sum_data=np.sum(data)
n=len(data)
mean=sum_data/n
print(mean)
```

**MEDIAN:**

```python
#This code is done by Jumana
median=np.median(data)
print("Median is :",median)
```

**MODE:**

```python
#This code is done by Jumana
def custom_mode(arr):
  unique_values,counts=np.unique(arr,return_counts=True)
  max_count=np.max(counts)
  mode = unique_values[counts == max_count]
  return mode
mode=custom_mode(data)
print("Mode:",mode)
```

**VARIANCE:**

```python
#This code is done by Jumana
variance=np.var(data)
print("Variance is:\n",variance)
```

**STANDARD DEVIATION:**

```python
#This code is done by Jumana
std=np.std(data)
print("Standard deviation is:\n",std)
```

**IQR:**

```python
#This code is done by Jumana
data_sorted=np.sort(data)
n=len(data_sorted)
q1_index=int(n*0.25)
q3_index=int(n*0.75)
q1=data_sorted[q1_index]
q3=data_sorted[q3_index]
iqr=q3-q1
print("Q1(25th percentile):", q1)
print("Q3(75th percentile):", q3)
print("IQR:", iqr)
```

**SKEWNESS:**

```python
#This code is done by Jumana
mean=data.mean()
std=np.std(data,ddof=1)   #ddof stands for degree of freedom which uses the sample standard deviation formula with n-1 as the divisor instead of the population standard deviation formula with n as the divisor.
n=len(data)
skewness=(np.sum((data-mean)**3)/n)/(std**3)
print("Skewness:\n",skewness)
```

**KURTOSIS:**

```python
#This code is done by Jumana
mean=data.mean()
std=np.std(data,ddof=1)   #ddof stands for degree of freedom which uses the sample standard deviation formula with n-1 as the divisor instead of the population standard deviation formula with n as the divisor.
n=len(data)
kurtosis=((np.sum((data-mean)**4)/n)/(std**4))-3
print("Kurtosis:\n",kurtosis)
```

**VERIFICATION USING SCIPY:**

```python
#This code is done by Jumana
mean=data.mean()
print("Mean after verification:\n",mean)
print("\n")
median=data.median()
print("Median after verification:\n",median)
print("\n")
mode=stats.mode(data,axis=None, keepdims=False)
print("Mode after verification:\n",mode)
print("\n")
variance=variation(data,ddof=1)
print("Variance after verification:\n",variance)
print("\n")
std=stats.tstd(data)
print("Standard deviation after verification:\n",std)
print("\n")
IQR=stats.iqr(data,interpolation='midpoint')
print("IQR after verification:\n",IQR)
print("\n")
skewness=data.skew()
print("Skewness after verification:\n",skewness)
print("\n")
kurtosis=data.kurt()
print("Kurtosis after verification:\n",kurtosis)
print("\n")
```

4. Calculate the values of skew and kurtosis using the general expressions discussed in the class and compare with the above results.

```python
#This code is done by Jumana
def skew(i):
  m=np.mean(i)
  sd=np.std(i)
  s=[]
  for x in i:
    s.append((x-m)**3)
  sk=np.mean(s)
  skew=sk/(sd**3)
  return skew
print("I:",skew(data['I']))
print("T:",skew(data['T']))
print("C:",skew(data['C']))
print("Q:",skew(data['Q']))
print("PF:",skew(data['PF']))
print("LF",skew(data['LF']))
```

```python
#This code is done by Jumana
def kurt(i):
 m=np.mean(i)
 sd=np.std(i)
 k=[]
 for x in i:
  k.append(((x-m)**4))
 ku=np.mean(k)
 kurtosis=(ku/(sd**4))-3
 return kurtosis
print("I:",kurt(data['I']))
print("T:",kurt(data['T']))
print("C:",kurt(data['C']))
print("Q:",kurt(data['Q']))
print("PF:",kurt(data['PF']))
print("LF:",kurt(data['LF']))
```

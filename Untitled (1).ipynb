```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
x=pd.read_excel("C:\\Users\\aidsa2\\Downloads\\stores.xlsx")
```

```python
x.info()
```

```python
x['shipment_days'] = (x['Ship Date'] - x['Order Date'])
```

```python
x.head(10)
```

```python
plt.figure(figsize=(9,7))
plt.hist(x["Quantity"],bins=20,color='aqua')
plt.xlabel("Quantity")
plt.ylabel("Frequency")
plt.title("HISTOGRAM")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(9,7))
plt.scatter(x["Quantity"],x["Profit"],alpha=0.34,color='slateblue')
plt.xlabel("Quantity")
plt.ylabel("Product")
plt.title("LIZA")
plt.grid(True)
plt.show()
```

```python
numeric_columns = x.select_dtypes(include='number')
correlation_matrix = numeric_columns.corr()

# Create the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()






```

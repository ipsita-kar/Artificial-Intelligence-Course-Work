```
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
# Define parameters
n = 10
p_values = [0.1, 0.3, 0.5, 0.7, 0.9]
# Plot PMF and CDF for each value of p
fig, axs = plt.subplots(len(p_values), 2, figsize=(10, 20))
for i, p in enumerate(p_values):
 # Calculate PMF and CDF using the expressions
 x = np.arange(0, n+1)
 pmf = stats.binom.pmf(x, n, p)
 cdf = stats.binom.cdf(x, n, p)
 # Plot PMF
 axs[i, 0].bar(x, pmf)
 axs[i, 0].set_title(f"p={p}")
 axs[i, 0].set_xlabel("x")
 axs[i, 0].set_ylabel("P(X=x)")
 # Plot CDF
 axs[i, 1].step(x, cdf)
 axs[i, 1].set_title(f"p={p}")
 axs[i, 1].set_xlabel("x")
 axs[i, 1].set_ylabel("P(Xâ‰¤x)")
plt.tight_layout()
plt.show()
```

```
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
# define the parameters
n = 10
p_values = [0.2, 0.5, 0.8]
# calculate the PMF and CDF for each value of p
for p in p_values:
 # calculate the PMF using the binomial distribution formula
 pmf = [stats.binom.pmf(k, n, p) for k in range(n+1)]

 # calculate the CDF using the cumulative distribution function

 cdf = [stats.binom.cdf(k, n, p) for k in range(n+1)]

 # plot the PMF and CDF
 fig, ax = plt.subplots(1, 2, figsize=(10, 4))
 ax[0].bar(range(n+1), pmf)
 ax[0].set_xlabel('k')
 ax[0].set_ylabel('P(X=k)')
 ax[0].set_title('PMF for p = {}'.format(p))

 ax[1].step(range(n+1), cdf)
 ax[1].set_xlabel('k')
 ax[1].set_ylabel('P(X<=k)')
 ax[1].set_title('CDF for p = {}'.format(p))

 plt.tight_layout()
 plt.show()
```

```
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
# Define parameters
mu_values = [1, 3, 5, 7, 9]
# Plot PMF and CDF for each value of mu
fig, axs = plt.subplots(len(mu_values), 2, figsize=(10, 20))
for i, mu in enumerate(mu_values):
 # Calculate PMF and CDF using the expressions
 x = np.arange(0, 21)
 pmf = stats.poisson.pmf(x, mu)
 cdf = stats.poisson.cdf(x, mu)
 # Plot PMF
 axs[i, 0].bar(x, pmf)
 axs[i, 0].set_title(f"mu={mu}")
 axs[i, 0].set_xlabel("x")
 axs[i, 0].set_ylabel("P(X=x)")
 # Plot CDF
 axs[i, 1].step(x, cdf)
 axs[i, 1].set_title(f"mu={mu}")
 axs[i, 1].set_xlabel("x")
 axs[i, 1].set_ylabel("P(Xâ‰¤x)")
plt.tight_layout()
plt.show()
```

```
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
# Define parameters
mu = 0
sigma_values = [0.5, 1, 1.5, 2, 2.5]
# Plot PDF and CDF for each value of sigma
x = np.linspace(-10, 10, 1000)
fig, ax = plt.subplots(figsize=(10, 6))
for sigma in sigma_values:
 # Calculate PDF and CDF using the normal distribution
 pdf = stats.norm.pdf(x, mu, sigma)
 cdf = stats.norm.cdf(x, mu, sigma)
 # Plot PDF and CDF
 ax.plot(x, pdf, label=f"sigma={sigma}")
 ax.plot(x, cdf, label=f"sigma={sigma}", linestyle="--")
# Set plot labels and legend
ax.set_xlabel("x")
ax.set_ylabel("f(x) / F(x)")
ax.set_title("PDF and CDF for Normal Distribution")
ax.legend()
plt.show()
```

```
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
from sklearn.preprocessing import StandardScaler
# Define parameters
mu, sigma = 0, 1
n = 10000
# Generate Normal random vector
x = np.random.normal(mu, sigma, n)
# Plot histogram of x
plt.hist(x, bins=50, density=True)
plt.title("Histogram of Normal Random Vector")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
# Convert x to a standard Normal distribution
scaler = StandardScaler()
x_standard = scaler.fit_transform(x.reshape(-1, 1)).flatten()
# Plot histogram of x_standard
plt.hist(x_standard, bins=50, density=True)
plt.title("Histogram of Standard Normal Random Vector")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

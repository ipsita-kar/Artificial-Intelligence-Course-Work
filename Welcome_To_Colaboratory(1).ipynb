```
import numpy as np
import matplotlib.pyplot as plt

# Define the PDF of X
def pdf_X(x):
    return 1 if 0 <= x <= 1 else 0

# Define the PDF of Y
def pdf_Y(y):
    return np.exp(-y) if y >= 0 else 0

# Define the PDF of Z
def pdf_Z(z):
    if z >= 0:
        result = 0
        for x in np.linspace(0, 1, 1000):  # Approximation using numerical integration
            y = z - x
            if y >= 0:
                result += pdf_X(x) * pdf_Y(y)
        return result
    else:
        return 0

# Generate data points for plotting
z_values = np.linspace(0, 5, 100)  # Range of z values
pdf_values = [pdf_Z(z) for z in z_values]  # Compute PDF values for each z

# Plot the PDF of Z
plt.plot(z_values, pdf_values, label='PDF of Z')
plt.xlabel('Z')
plt.ylabel('Probability Density')
plt.title('PDF of Z = X + Y')
plt.legend()
plt.grid(True)
plt.show()
```
